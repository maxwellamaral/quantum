## Experimento de Interferência com Hadamard (Interferômetro de Mach-Zehnder)

Você pode replicar esse experimento físico agora mesmo no Qiskit. É a prova de que a porta Hadamard é reversível.

Matematicamente, aplicar dois Hadamards é igual a não fazer nada (Identidade), pois as superposições se cancelam e voltam ao estado original.

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
```

```python

qc = QuantumCircuit(1)

# Passo 1: O fóton encontra o primeiro espelho
# Estado: Superposição (50/50)
qc.h(0) 
```

### Visualização após o primeiro Hadamard (Espelho Semitransparente 1)

```python
# Esfera de Bloch após o primeiro H
vetor_apos_h1 = Statevector(qc)
display(plot_bloch_multivector(vetor_apos_h1))
print(f"Estado após primeiro H: {vetor_apos_h1}")
```

```python
# Q-Sphere após o primeiro H
display(plot_state_qsphere(vetor_apos_h1))
```

```python

# Passo 2: O fóton viaja pelos caminhos e encontra o SEGUNDO espelho
# A interferência acontece aqui.
qc.h(0)
```

### Visualização após o segundo Hadamard (Espelho Semitransparente 2 - Interferência)

```python
# Esfera de Bloch após o segundo H
vetor_apos_h2 = Statevector(qc)
display(plot_bloch_multivector(vetor_apos_h2))
print(f"Estado após segundo H: {vetor_apos_h2}")
```

```python
# Q-Sphere após o segundo H
display(plot_state_qsphere(vetor_apos_h2))
```

### Visualização do Circuito Completo

```python

# Verificação
vetor = Statevector(qc)
print(vetor)
```

```python
# Desenho do circuito quântico
qc.draw('mpl')
```
